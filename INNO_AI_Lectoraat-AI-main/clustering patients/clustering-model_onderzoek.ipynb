```python
import sklearn.cluster as cl
import sklearn.mixture as mi
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
df = pd.read_csv("DES.csv")
df
```

```python
dfx = df.drop(columns=['Unnamed: 0', "patient_id", "geboortedatum", "opname_nummer", "zkh_opn_start", "icd_version", "start_operatie", "endtime", "duur_ok_minutes", "length_of_stay_days", "time_to_surgery_minutes", "target", "diagnose"])
dfy = df.drop(columns=['Unnamed: 0'])

dfx
```

```python
df_set = []

for column in dfx.columns:
    df_set.append(pd.get_dummies(dfx[column], prefix = column))
    print(f"encoded {column} : {len(dfx[column].unique())}")

df_hot = pd.DataFrame()
for set in df_set:
    df_hot = pd.concat([df_hot, set], axis=1)
    # print(df_hot)
print("Made single set")



# dfx["herkomst"] = dfx["herkomst"].astype('category').cat.codes
# dfx["marital_status"] = dfx["marital_status"].astype('category').cat.codes
# dfx["race"] = dfx["race"].astype('category').cat.codes
# dfx["diagnose"] = dfx["diagnose"].astype('category').cat.codes
# dfx["department"] = dfx["department"].astype('category').cat.codes

df_hot
```
