```python
import pandas as pd
```

```python
stays = pd.read_csv("icustays.csv")
stays = stays.drop(columns=["first_careunit", "last_careunit"])
stays["intime"] = pd.to_datetime(stays["intime"])
stays["outtime"] = pd.to_datetime(stays["outtime"])
stays = stays.rename(columns={"intime": "icu_in", "outtime": "icu_out"})
stays
```

```python
admiss = pd.read_csv("admissions.csv")
admiss = admiss.drop(columns=["admission_type", "admit_provider_id", "hospital_expire_flag", "insurance", "language", "marital_status", "race", "deathtime", "edregtime", "edouttime",
                               "discharge_location"])
admiss["admittime"] = pd.to_datetime(admiss["admittime"])
admiss["dischtime"] = pd.to_datetime(admiss["dischtime"])
admiss
```

```python
diagnose  = pd.read_csv("d_icd_diagnoses.csv")
```

```python
diagnose_icd = pd.read_csv("diagnoses_icd.csv")
diagnose_icd["diagnose"] = pd.merge(diagnose, diagnose_icd, on=['icd_code', 'icd_version'], how='right')['long_title']
diagnose_icd = diagnose_icd[diagnose_icd['seq_num'] <= 1]
diagnose_icd = diagnose_icd.reset_index(drop=True)
diagnose_icd
```

```python
procedure = pd.read_csv("d_icd_procedures.csv")
procedure
```

```python
procedure_icd = pd.read_csv("procedures_icd.csv")
procedure_icd["procedure"] = pd.merge(procedure, procedure_icd, on=['icd_code', 'icd_version'], how='right')['long_title']
procedure_icd = procedure_icd[procedure_icd['seq_num'] <= 1]
procedure_icd = procedure_icd.reset_index(drop=True)
procedure_icd
```

```python
omr = pd.read_csv("omr.csv")
omr["chartdate"] = pd.to_datetime(omr["chartdate"])
omr = omr[omr['result_name'].isin(['BMI (kg/m2)','BMI'])]
omr = omr.sort_values(by=["subject_id", "chartdate"], ascending=False).drop_duplicates(subset=['subject_id']).sort_values(by="subject_id")
omr = omr.reset_index(drop=True)
omr
```
